```python
from scipy.interpolate import interp1d
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Linear Interpolasi

```python
 x = [0, 1, 2]
 y = [1, 3, 2]
```

```python
f = interp1d(x, y)
y_ask = f(1.5)
print(y_ask)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(1.5, y_ask, 'ro')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Linear Interpolasi Newton

```python
%matplotlib inline
def divided_diff(x, y):
  n = len(y)
  coef = np.zeros([n,n])
  # the first column is y
  coef[:,0] = y

  for j in range(1,n):
    for i in range(n-j):
      coef[i][j] = \
      (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
  return coef

def newton_poly(coef, x_data, x):
  n = len(x_data) - 1
  p = coef[n]
  for k in range(1,n+1):
    p = coef[n-k] + (x -x_data[n-k])*p
  return p
```

```python
x = np.array([-5, -1, 0, 2])
y = np.array([-2, 6, 1, 3])

a_s = divided_diff(x, y)[0, :]

x_new = np.arange(-5, 2.1, .1)
y_new = newton_poly(a_s, x, x_new)

plt.figure(figsize = (12,8))
plt.plot(x, y, 'bo')
plt.plot(x_new, x_new, 'go')
plt.plot(x_new, y_new)
```

```python
jarak = np.array([0, 5, 10])
kecepatan = np.array([10, 15, 20])
dicari = 7.2
```

```python
kec_target = np.interp(dicari, jarak, kecepatan)
print(f"Kecepatan pada jarak {dicari} meter adalah {dicari:.2f} m/detik")

plt.figure(figsize = (12,8))
plt.plot(jarak, kecepatan, 'bo-')
plt.plot(dicari, kec_target, 'ro')
plt.xlabel('jarak')
plt.ylabel('kecepatan')
plt.title("Interpolasi Linear Ben Johnson")
plt.show()
```
