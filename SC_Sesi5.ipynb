# Materi Kelas

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-poster")
%matplotlib inline

h = 0.2
x = np.arange(0, 2*np.pi, h)
y = np.cos(x)

forward_diff = np.diff(y)/h
x_diff = x[:-1]

exact_solution = -np.sin(x_diff)

plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', label = "Finite Difference Approx")
plt.plot(x_diff, exact_solution, label = "Exact Solution")
plt.legend()
plt.show()

max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

```python
#simpsons rule
a = 0
b = np.pi
n = 10
h = (b-a)/n
x = np.linspace(a, b, n+1)
y = np.sin(x)

S = h/3 * (y[0]+y[-1] + 4 * sum(y[1:-1:2]) + 2 * sum(y[2:-2:2]))
exact = 2
error = abs(S-exact)
print("simpsons: ", S)
print("Exact: ", exact)
print("error: ", S-exact)
```

```python
import numpy as np
from math import sin, pi

x = np.array([0, 0.1, 0.2, 0.3, 0.4])
y = np.array([0, 0.0819, 0.1341, 0.1646, 0.1797])
h = 0.1

def turunan_pertama(y, h, index):
    return (y[index + 1] - y[index]) / h

dy_dx_x0 = turunan_pertama(y, h, 0)
dy_dx_x02 = turunan_pertama(y, h, 2)

print("Turunan pertama untuk (x = 0):", dy_dx_x0)
print("Turunan pertama untuk (x = 0.2):", dy_dx_x02)
```

```python
def f(x):
    return np.sin(x)

def trapezoid(a, b, n):
    x = np.linspace(a, b, n + 1)
    y = f(x)
    h = (b - a) / n
    integral = h * (y[0] + y[-1] + 2 * np.sum(y[1:-1])) / 2
    return integral

a = 0
b = pi

result_8 = trapezoid(a, b, 8)
result_16 = trapezoid(a, b, 16)

print("Hasil 8 panel:", result_8)
print("Hasil 16 panel:", result_16)
```
